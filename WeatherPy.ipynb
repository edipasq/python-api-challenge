```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress
import json
from config import weather_key
api_key = weather_key
from citipy import citipy
from datetime import date


lat_range = (-90, 90)
lng_range = (-180, 180)

url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"


query_url = f"{url}appid={api_key}&units={units}&q="
```

```python
lat_lngs = []
cities = []


lats = np.random.uniform(lat_range[0], lat_range[1], size=1500)
lngs = np.random.uniform(lng_range[0], lng_range[1], size=1500)
lat_lngs = zip(lats, lngs)


for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
   
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

```python
cities
```

```python
lat = []
long = []
temp = []
Cloudiness = []
Wind = []
Date = []
Country = []
Humidity = []
Cities_Found = []
City_id = []
num = 1
# Loop through the list of cities and perform a request for data on each
for city in cities:

    response = requests.get(query_url + city).json()
    try:
        
        lat.append(response['coord']['lat'])
        long.append(response['coord']['lon'])
        temp.append(response['main']['temp_max'])
        Wind.append(response['wind']['speed']) 
        Date.append(response['dt']) 
        Country.append(response['sys']['country']) 
        Humidity.append(response['main']['humidity']) 
        Cloudiness.append(response['clouds']['all']) 
        Cities_Found.append(city)
        print(f"{city} ---- found! CityNumber:{num} Record processing")
        City_id.append(num)
        num = num + 1
    except:
        print(f"{city} ----city not found")
        pass
```

```python
len(City_id)

```

```python
City_Weather = pd.DataFrame({
    'City Id': City_id,
    'City'   : Cities_Found,
    'Cloudiness': Cloudiness,
    'Country':Country,
    'Date':Date,
    'Humidity':Humidity,
    'lattitude'    : lat,
    'long'   : long, 
    'Max temp'   : temp, 
    'Wind Speed' : Wind
})
City_Weather
```

```python
title1=(f"{Date1} ----Northen Hemisphere Temp VS Latitude")
x_values = Northen_Hemisphere['lattitude']
y_values = Northen_Hemisphere['Max temp']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(10,30),fontsize=8,color="red")
plt.title(title1)
plt.xlabel('Latitude')
plt.ylabel('Temperature')
print(f"The r-value is: {rvalue**2}")
plt.savefig("output_data/NHTvL.png")
plt.show()
```
